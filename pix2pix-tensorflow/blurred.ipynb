```python
import os
import PIL.Image
from PIL import ImageFilter

def blur_images():
    rootdir = 'resized/'
    for _, _, files in os.walk(rootdir):
        for file in files:
            if 'png' in file:
                print(file)
                img = rootdir + file
                im = PIL.Image.open(img)
                im1 = im.filter(ImageFilter.GaussianBlur(radius=5))
                im1.save('blurred_5/' + file)
```
